```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
house_price["Condition"]=le.fit_transform(house_price["Condition"])
house_price.head(6)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
house_price["Location"]=le.fit_transform(house_price["Location"])
house_price.head(6)
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Define features (X) and target (y)
x = house_price[['Area','Bedrooms','Bathrooms','Floors','YearBuilt']]
y = house_price['Price']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
# make prediction using the test data
y_pred=model.predict(X_test)
```

```python
plt.scatter(X_test['Area'], y_test, color='b', label='Actual Data')
plt.plot(X_test['Area'], y_pred, color='r', label='Regression Line')
plt.xlabel('Area')
plt.ylabel('Price')
plt.legend()
plt.title('House Price Prediction Using Linear Regression')
plt.show()
```
